Tier 3: Original Design

Title: "Blind, Deaf, and Uneducated"

This is an idea where I thought about accessibility to all people, regardless of what disability you may have, and or if you understand Morse code.

So, I have these components:

1x OLED Screen - for displaying what message is currently being communicated to the user via the buzzer and LED

1x LED (was originally white, but was too bright) - This is what shows the actual morse code in the flashes of light

1x Buzzer - for those who are blind, but able to listen to what the message is from the morse code in sound form

Now of course, I had Gemini make the code because OLED screens are a bit complicated to work with, and the logic itself would take a long time to understand if I brute forced it.

The actual building of the circuit itself in real life was done by me, since it isn't complicated whatsoever. The only real danger is using the LED without the 220 Ohm resistor, and even then, that's simple enough to avoid.

Gemini prompt:

"I have an idea for an arduino project that I want to do, but I'd like some assistance in developing it.

The project itself is going to be related to people with disabilities like being deaf, and blind.

These are the components I will be using.

An OLED screen to display messages for those who are deaf, but unable to understand blinking lights for morse code.

A buzzer for the blind to listen to the same message but in morse code.

An LED for those who are deaf but can understand morse code blinking."
